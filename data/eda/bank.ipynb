```python
df["age"] = pd.qcut(df["age"], q=4, labels=False, duplicates="drop")
df["balance"] = pd.qcut(df["balance"], q=4, labels=False, duplicates="drop")
df["day"] = pd.qcut(df["day"], q=4, labels=False, duplicates="drop")
df["duration"] = pd.qcut(df["duration"], q=4, labels=False, duplicates="drop")
df["campaign"] = pd.qcut(df["campaign"], q=4, labels=False, duplicates="drop")
df["pdays"] = pd.qcut(df["pdays"], q=4, labels=False, duplicates="drop")
df["previous"] = pd.qcut(df["previous"], q=4, labels=False, duplicates="drop")
```

```python
df_enc = df.apply(lambda s: s.astype('category').cat.codes)
```
